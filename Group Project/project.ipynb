```python
X = data.iloc[:,1:].values
Y = data.iloc[:,0].values
Y
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
std_data = ss.fit_transform(np.c_[Y, X])
#std_X = np.c_[np.ones([len(y)]).T, std_data[:,1:11]]
std_X = std_data[:,1:]
std_Y = std_data[:,0]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt

#定义模型
model = LinearRegression()

# 初始化用于存储MSE和MAE的列表
train_mse = []
train_mae = []
test_mse = []
test_mae = []

# 定义10折交叉验证
kfold = KFold(n_splits=5, shuffle=True, random_state=999)

# 进行交叉验证并计算每次验证的特征重要度得分
for train_index, test_index in kfold.split(X):
    X_train, X_test = std_X[train_index], std_X[test_index]
    y_train, y_test = std_Y[train_index], std_Y[test_index]
    
    # 训练模型
    model.fit(X_train, y_train)
    
    # 使用模型进行训练集和测试集的预测
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # 存储到列表中
    train_mse.append(mean_squared_error(y_train, y_train_pred))
    train_mae.append(mean_absolute_error(y_train, y_train_pred))
    test_mse.append(mean_squared_error(y_test, y_test_pred))
    test_mae.append(mean_absolute_error(y_test, y_test_pred))

print("Train MSE scores:", train_mse)
print("Test MSE scores:", test_mse)
print("Train MAE scores:", train_mae)
print("Test MAE scores:", test_mae)

total_train_mae_mean = np.mean(train_mae)
total_train_mse_mean = np.mean(train_mse)
total_test_mae_mean = np.mean(test_mae)
total_test_mse_mean = np.mean(test_mse)
total_dict = {'total_train_mae_mean':total_train_mae_mean, 'total_train_mse_mean':total_train_mse_mean, 'total_test_mae_mean':total_test_mae_mean, 'total_tset_mse_mean':total_test_mse_mean}
print(total_dict)
```

```python
#画图
x_axis=np.linspace(1,5,5)
plt.rcParams["figure.figsize"] = (18,18)
plt.plot(x_axis,test_mse,"g-",linewidth=3,label="testing MSE")
plt.plot(x_axis,test_mae,"y-",linewidth=3,label="testing MAE")
plt.plot(x_axis,train_mse,"b-",linewidth=3,label="training MSE")
plt.plot(x_axis,train_mae,"r-",linewidth=3,label="training MAE")
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.legend(loc="upper left",fontsize=30)
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

#定义模型
model = RandomForestRegressor(n_estimators=200,random_state=42)

# 初始化用于存储MSE和MAE的列表
train_mse = []
train_mae = []
test_mse = []
test_mae = []

# 初始化一个数组来存储每次验证的特征重要度得分
feature_importance_scores = []

# 定义10折交叉验证
kfold = KFold(n_splits=5, shuffle=True, random_state=999)

# 进行交叉验证并计算每次验证的特征重要度得分
for train_index, test_index in kfold.split(X):
    X_train, X_test = std_X[train_index], std_X[test_index]
    y_train, y_test = std_Y[train_index], std_Y[test_index]
    
    # 训练模型
    model.fit(X_train, y_train)
    
    # 使用模型进行训练集和测试集的预测
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # 存储到列表中
    train_mse.append(mean_squared_error(y_train, y_train_pred))
    train_mae.append(mean_absolute_error(y_train, y_train_pred))
    test_mse.append(mean_squared_error(y_test, y_test_pred))
    test_mae.append(mean_absolute_error(y_test, y_test_pred))

    # 获取特征重要度得分并存储
    feature_importance_scores.append(model.feature_importances_)

print("Train MSE scores:", train_mse)
print("Test MSE scores:", test_mse)
print("Train MAE scores:", train_mae)
print("Test MAE scores:", test_mae)

total_train_mae_mean = np.mean(train_mae)
total_train_mse_mean = np.mean(train_mse)
total_test_mae_mean = np.mean(test_mae)
total_test_mse_mean = np.mean(test_mse)
total_dict = {'total_train_mae_mean':total_train_mae_mean, 'total_train_mse_mean':total_train_mse_mean, 'total_test_mae_mean':total_test_mae_mean, 'total_tset_mse_mean':total_test_mse_mean}
print(total_dict)
```

```python
# 输出每次验证的特征重要度得分
for i, score in enumerate(feature_importance_scores):
    print(f"第{i+1}次验证的特征重要度得分:")
    indices = np.argsort(score)[::-1]
    for f in range(X.shape[1]):
        print("%2d) %-*s %f" % (f + 1, 30, feature[indices[f]], score[indices[f]]))
```
